```python
class Content:
    args = ''
    speaker = ''
    msg = ''
    cleaned_msg = ''
    loc = []
```

```python
import os
def get_files():
    path = r'C:\Users\ThinkPad\Desktop\fgostory'
    return os.listdir(path)
```

```python
def msg_clean(msg):
    msg = msg.replace('[r]','\n')
    msg = msg.replace('[line 3]','——————')
    msg = msg.replace('[line 6]','————————————')
    msg = msg.replace('[%1]','藤丸立香')
    msg = msg.replace('[&他:她]','他')
    for space in ['[-]','[51d4ff]','[k]']:
        msg = msg.replace(space,'')
    return msg
```

```python
def read_file(fname):
    path = r'C:\Users\ThinkPad\Desktop\fgostory'
    file = open(os.path.join(path,fname),encoding = 'utf-8')
    story = file.read()
    return story.split('\n\n')
```

```python
def handle_file(story_split):
    story_loc = story_split[0][1:].split('-')
    real_story = []
    for c in story_split:
        c = c.strip()
        if c.find('＠') != -1 or c.find('？') == 0:
            real_story.append(c)

    story_container = []
    for r in real_story:
        temp = Content()
        speaker_start = r.find('＠')
        if speaker_start == -1:
            temp.speaker = 'Master选择'
            temp.args = ''
            if r == '？！':
                temp.speaker = ''
                temp.msg = '『选择结束』\n'
            else:
                temp.msg = r[1:] + '\n'
        else:
            speaker_end = r.find('\n',speaker_start)
            temp.args = r[:speaker_start]
            temp.speaker = msg_clean(r[speaker_start + 1:speaker_end])
            temp.msg = r[speaker_end:].strip()
        temp.cleaned_msg = msg_clean(temp.msg)
        temp.loc = story_loc
        story_container.append(temp)
    return story_container
```

```python
def publish_story(story_container):
    output = ''
    current_speaker = ''
    for sc in story_container:
        if current_speaker == sc.speaker or not sc.speaker:
            pass
        else:
            output += f'『{sc.speaker}』\n\n'
        current_speaker = sc.speaker
        output += sc.cleaned_msg +'\n'
    return output
```

```python
file_list = get_files()
for f in file_list[:10]:
    sf = read_file(f)
    sc = handle_file(sf)
    ss = publish_story(sc)
```

```python
path = r'C:\Users\ThinkPad\Desktop\fgostory_clean'
for i in range(len(file_list)):
    sf = read_file(file_list[i])
    sc = handle_file(sf)
    sp = publish_story(sc)
    with open(os.path.join(path,file_list[i]),'w+',encoding='utf-8') as f:
        f.write(sp)
    print(f'{i}/5856',end=' ')
```

```python
clean_path = r'C:\Users\ThinkPad\Desktop\fgostory_clean'
sum_path = r'C:\Users\ThinkPad\Desktop\fgostory_sum'
clean_files = os.listdir(clean_path)
current_chap = '00'
current_story = ''
for file in clean_files:
    file_chap = file[6:8]
    file_content = open(os.path.join(clean_path,file),encoding='utf-8')
    file_story = file_content.read()
    if file_chap == current_chap:
        current_story += file_story
    else:
        with open(os.path.join(sum_path,file[:8] + '.md'),'w+',encoding='utf-8') as f:
            f.write(current_story)
        current_chap = file_chap
        current_story = file_story
    file_content.close()
print('over')

```
